# 3a.DATA CLEANING

Data cleaning basically highlights the unwanted portion of the data and remove the impurities.

This stage involves;

removing unwanted characters (e.g., emojis, placeholders) 
2.Handling missing data (no missing data found in the dataset)
3.Normalizing text 
(e.g., lowercasing), Remove stopwords, Remove Punctuation

```python
import re # Using Regular expression module
import nltk
import string
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
punctuation = string.punctuation  # includes !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~

# chage all uppercase text ti lowercase, create a new column for lowercase text, now, df['text_lower'] contains the lowercased version of each tweet.
df['text_lower'] = df['text'].str.lower()

df[['text', 'text_lower']].head() 
#this displays the first five data in the dataset, showing original vs. lowercased tweets side-by-side.

def clean_tweet(text):
    # Lowercase
    text = text.lower()

    # Remove tokenized mentions and placeholders
    text = re.sub(r'\{@.*?@\}', '', text)
    text = re.sub(r'\{\{url\}\}', '', text)
    text = re.sub(r'\{\{username\}\}', '', text)

    # Remove emojis and non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)

    # Remove punctuation (but keep hashtags)
    text = re.sub(f"[{re.escape(punctuation.replace('#',''))}]", "", text)

    # Remove #  from hastags but let the words after it remain
    text = re.sub(r'#(\w+)', r'\1', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove stopwords
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]

    return ' '.join(tokens)
    #Textprocessing_stem function using porter streamer pre-processor
def Textprocessing_stem(text):
    #renove devanagart script
    text = re.sub(r' |\u0900-\u0976|+', ' ', text)
    
```

# Lemmatization Section

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet, stopwords
from nltk import pos_tag

# Download required resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

stop_words = set(stopwords.words('english'))

# Function to map POS tags to WordNet tags
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default to noun

# Preprocessing function: tokenize + lemmatize
def preprocess_text(text):
    try:
        tokens = word_tokenize(text)
        pos_tags = pos_tag(tokens)
        lemmatized = [
            lemmatizer.lemmatize(token, get_wordnet_pos(tag))
            for token, tag in pos_tags
            if token.lower() not in stop_words
        ]
        return ' '.join(lemmatized)
    except Exception as e:
        print(f"Error: {e}")
        return text

# Apply to your cleaned text column
df['text_lemmatized'] = df['text_clean'].apply(preprocess_text)

# Preview
df[['text_clean', 'text_lemmatized']].head()
```

# STEMMING Section

Investing the stemming method for breaking words down to their root forms in addition to lemmatization. 

```python
import time
import pandas as pd
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

# Define stemming function
def Textprocessing_stem(text):
    tokens = text.split()
    stems = [stemmer.stem(word) for word in tokens]
    return ' '.join(stems)

# Set max column width for better display
pd.set_option('display.max_colwidth', 100)

# Copy the original DataFrame
df_cleaned = df.copy()

# Apply stemming to cleaned text
start_time = time.time()
df_cleaned['text_stemmed'] = df_cleaned['text_clean'].apply(Textprocessing_stem)
process_time = time.time() - start_time

print("Time taken for data processing with stemming preprocessing:", round(process_time, 2), "seconds")

# Show a preview
df_cleaned[['text_clean', 'text_stemmed']].head()

```
